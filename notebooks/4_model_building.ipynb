```python
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies = pd.read_csv("../data/processed/final_movies.csv")
```

```python
movies.head()
```

Vectorizing the words, since machine understands numbers not letters.

```python
cv = CountVectorizer(max_features=5000, stop_words="english")
```

```python
vectors = cv.fit_transform(movies["tags"])
```

```python
vectors.shape #(movies, features)
```

```python
vectors = vectors.toarray() #sparse conversion to array
```

```python
vectors[0]
```

```python
#cosine similarity to compare movies
#Range = 0 to 1(higher=similar, 1=exact same)
similarity = cosine_similarity(vectors) 
```

```python
similarity[0][0]
```

```python
similarity.shape #Movies V/S Movies
```

```python
#Recommendation function
def recommend(movie):
    #find index
    movie_index = movies[movies["title"]== movie].index[0]
    
    #get similarities
    distances = similarity[movie_index]
    
    #sort the similarity values(highest similar first)
    movies_list = sorted(enumerate(distances), reverse=True, key=lambda x: x[1])
    
    #skip first(AVATAR==AVATAR😁) and check other 5
    for i in movies_list[1:6]:
        print(movies.iloc[i[0]].title)
```

```python
recommend("Avatar")
```
